```python
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from astroNN.datasets import galaxy10
```

```python
images, labels = galaxy10.load_data()
```

```python
np.where(labels == 0)[0]  # Check the number of images in class 0
```

```python
images_all = images
images_50 = images[0:1000]
```

```python
norm_images = images_50.astype('float32') / 255.0 # Normalize the images to [0, 1]
```

```python
images_50.shape
```

```python
few_shot_size = 50
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(norm_images, labels[0:1000], train_size=few_shot_size, random_state=42)
```

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_train[i])
    ax.axis("off")
plt.show()
```
